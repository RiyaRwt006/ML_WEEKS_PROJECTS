– Introduction

The project explores risk factors of heart disease.

Dataset has 303 patients with 14 attributes such as age, gender, cholesterol, chest pain, etc.

Goal: Find patterns that explain the presence or absence of heart disease.

 - Demographic Patterns

Age: Most patients fall between 40 to 60 years old.

Gender: Men are slightly more affected than women.

This shows middle-aged males are at higher risk.

- Medical Risk Factors

Chest Pain Type (cp): Certain chest pain types (especially type 2 & 3) strongly indicate heart disease.

Cholesterol (chol): High cholesterol is common among patients with disease.

Max Heart Rate (thalach): Patients with lower heart rate are more likely to have disease.

ST Depression (oldpeak): Higher depression values are linked with heart disease.

 – Correlation Insights

Positive correlation with heart disease: ST depression (oldpeak), number of vessels (ca).

Negative correlation with heart disease: Maximum heart rate (thalach), chest pain type (cp).

Other factors like blood pressure and fasting blood sugar show weaker influence.

 – Outlier Observations

A few patients had cholesterol > 500 mg/dl → very unusual, possibly rare cases.

Some patients had extremely high blood pressure → important to monitor but less extreme.

– Final Summary

Key factors influencing heart disease:
-> Age, Gender, Chest Pain, Cholesterol, Max Heart Rate.

This analysis can guide predictive models and help doctors identify high-risk patients early.

```python
# Univariate Analysis
#Age distribution
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df['age'],bins=20 , kde=True,color="skyblue")
plt.title("Age distribution of patients")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
#Gender Distribution
plt.figure(figsize=(6,4))
sns.countplot(x='sex',data=df,palette='pastel')
plt.title("Gender distribution (0=Female,1=Male)")
plt.xlabel("Sex")
plt.ylabel("count")
plt.show()
```

```python
# Chest Pain
plt.figure(figsize=(6,4))
sns.countplot(x='cp',data = df,hue='cp', palette="muted",legend=False)
plt.title("Chest Pain Distribution")
plt.xlabel("Chest pain type (0-3)")
plt.ylabel("count")
plt.show()
```

```python
#Cholestrol Level

plt.figure(figsize=(8,5))
sns.histplot(df['chol'],bins=30,kde=True,color="orange")
plt.title("Cholestrol Distribution")
plt.xlabel("Cholestrol (mg/dl)")
plt.ylabel("count")
plt.show()
```

```python
#Target  (Heart Diseaseor Not)

plt.figure(figsize=(6,4))
sns.countplot(x='target',data=df,hue="target",palette='coolwarm',legend=False)
plt.title("Heart Disease Presence (0=No,1=Yes)")
plt.xlabel("Target")
plt.ylabel("Count")
plt.show()
```

```python
#Bivariate Analysis
#Age vs heart Diseases
plt.figure(figsize=(8,5))
sns.histplot(data=df, x="age", hue="target", kde=True, palette="Set2", bins=20)
plt.title("Age vs Heart Disease")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
#Gender vs Heart Diseases
plt.figure(figsize=(6,4))
sns.countplot(x="sex", hue="target", data=df, palette="coolwarm")
plt.title("Gender vs Heart Disease (0=Female, 1=Male)")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.legend(["No Disease", "Disease"])
plt.show()
```

```python
#chest pain type vs heart Diseases
plt.figure(figsize=(7,5))
sns.countplot(x="cp", hue="target", data=df, palette="muted")
plt.title("Chest Pain Type vs Heart Disease")
plt.xlabel("Chest Pain Type (0–3)")
plt.ylabel("Count")
plt.show()
```

```python
#Cholerstrol vs heart Diseases
plt.figure(figsize=(8,5))
sns.boxplot(x="target", y="chol",hue="target", data=df, palette="Set3",legend=False)
plt.title("Cholesterol Levels vs Heart Disease")
plt.xlabel("Target (0=No, 1=Yes)")
plt.ylabel("Cholesterol (mg/dl)")
plt.show()
```

```python
#Max Heart Rate vs Heart Disease
plt.figure(figsize=(8,5))
sns.boxplot(x="target", y="thalach", hue="target", data=df, palette="Set2",legend=False)
plt.title("Max Heart Rate vs Heart Disease")
plt.xlabel("Target (0=No, 1=Yes)")
plt.ylabel("Max Heart Rate")
plt.show()
```

```python
#Correlation Heatmap
plt.figure(figsize=(12,8))
corr = df.corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Correlation Heatmap")
plt.show()
```
